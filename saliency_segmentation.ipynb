```python
""" Use a grab-cut based method to segment out buildings.

Created on Sat May 11 16:07:02 2013

:author: John Femiani {john.femiani@asu.edu}
:author: Er Li
:author: Kunal Khanna

"""

import numpy as np
import cv2
import scipy.ndimage
import logging
from logging import info  # analysis:ignore
import math


def gs2rgb(gray_value):
    """ Convert grayscale value `a` to an RGB triple. """
    return np.dstack((gray_value, gray_value, gray_value))


def construct_structuring_element(kernel_size, LIGHT_VEC):
    """
    construct structure element from light vector with width as kernel_size
    """

    width = 2 * kernel_size + 1
    structures = np.zeros([width, width])

    if math.fabs(LIGHT_VEC[1]) == math.fabs(LIGHT_VEC[0]):
        xend = (LIGHT_VEC[0] > 0) and kernel_size or -1 * kernel_size
        yend = (LIGHT_VEC[1] > 0) and kernel_size or -1 * kernel_size
    else:
        if math.fabs(LIGHT_VEC[1]) > math.fabs(LIGHT_VEC[0]):
            yend = (LIGHT_VEC[1] > 0) and kernel_size or -1 * kernel_size
            xend = LIGHT_VEC[0] * yend / LIGHT_VEC[1]
        else:
            xend = (LIGHT_VEC[0] > 0) and kernel_size or -1 * kernel_size
            yend = LIGHT_VEC[1] * xend / LIGHT_VEC[0]
    xend = kernel_size + int(round(xend))
    yend = kernel_size + int(round(yend))

    cv2.line(structures, (kernel_size, kernel_size), (xend, yend), 1.0)

    return structures


def erase_small_shadows(shdows, threshold):
    """
    This is used before segmenting to remove speckles from
    the mask used by grabcut.

    Occasional small black specs can result from rocks or bushes
    or image compression artifacts, the GC algorithm proceeds better
    if they are pruned out early.

    :param shadows: An image that is nozero wherever a pixel is in shadow.
    :param threshold: The minimum shadow area to consider for grabcut.
    """

    contours, _ = cv2.findContours(shdows.astype(np.uint8),
                                   mode=cv2.RETR_EXTERNAL,
                                   method=cv2.CHAIN_APPROX_SIMPLE)
    small_contours = [c for c in contours if cv2.contourArea(c) < threshold]

    if len(small_contours) > 0:
        cv2.drawContours(shdows, small_contours, -1, color=0, thickness=-1)


def erase_thin_shadows(shadows, light_vec):
    """
    Remove shadows cast by short objects.

    A building is typically more than 3-4 meters high, whereas a wall or
    cars is unlikely to be more than 2-2.5. hese things cast shadows but
    the shadows are lileky shorter than the ones cast by buildings.

    Parameters
    ----------
    :shadows:  Used for both input and output. A mast that is nonzero
        wherever there is a shadows


    """
    # TODO:  Use erosion to get rid odf thin shadows
    # see OK et al.: DETECTION OF ARBITRARILY SHAPED BUILDINGS IN COMPLEX
    #               ENVIRONMENTS

    kernel_size = 3
    structuring_element = construct_structuring_element(kernel_size, light_vec)

    building_shadows = \
        scipy.ndimage.binary_opening(shadows,
                                     structure=structuring_element)

    return building_shadows


def mark_from_shadows(gs,
                      prob_roof,
                      median_blurred,
                      light_vec,
                      shadow_threshold,
                      ignore_thin_shadows,
                      fgd_distance):
    """
    Sets certain pixels prob_roof to BKG or FGD based on shadows.


    Parameters
    ----------
    :gs:  The intensities of the input aerial image
    :prob_roof: An image that is 0 for not-roof, 1 for roof, and in-between
        when the label is uncertain.
    :median_blurred: An image that is a median blurred version of the input (`gs`)

    """

#    from scipy.ndimage import gaussian_laplace
#    LoG = gaussian_laplace(gs, 2.0)
    if median_blurred is None:
        gs = cv2.medianBlur(gs, 3)
    else:
        gs = median_blurred

    shadows = (gs < shadow_threshold) & (prob_roof > 0)
    shadows = shadows.astype(gs.dtype)

    erase_small_shadows(shadows, 5)

    building_shadows = shadows

    if ignore_thin_shadows:
        building_shadows = erase_thin_shadows(shadows, light_vec)

    height, width = gs.shape  # rows = height, cols=width.....
    print "fgd_distance", fgd_distance
#    structures = construct_structuring_element(fgd_distance, -1.0 * light_vec)
#    shifted_shadows = scipy.ndimage.binary_dilation(
#        building_shadows,
#        structure=structures)
#    shifted_shadows = shifted_shadows & (~building_shadows.astype(np.uint8))
#    prob_roof[(shifted_shadows > 0) & (prob_roof > 0)] = 1.0
    
    
    structures = construct_structuring_element(fgd_distance, -1.0 * light_vec)
    shifted_shadows = scipy.ndimage.binary_dilation(
        building_shadows,
        structure=structures)
    shifted_shadows = shifted_shadows & (~building_shadows.astype(np.uint8))
    prob_roof[(shifted_shadows > 0) & (prob_roof > 0)] = 1.0

    prob_roof[(shadows > 0) & (prob_roof < 1)] = 0.0


def mark_trees(image, prob_image, tree_green_percent):
    """
    Mark vegetation as background based on the amount of green

    Parameters
    --------
    :image:   RGB image
    :prob_image:   confidence values


    """
    blurred_image = cv2.blur(image, (3, 3))
    trees = blurred_image[:, :, 1] > (1 + tree_green_percent / 100.0)\
        * np.max([blurred_image[:, :, 0], blurred_image[:, :, 2]], 0)
    trees = trees.astype(np.uint8)
    erase_small_shadows(trees, 10)
    trees = cv2.dilate(trees,
                       cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5)))
    prob_image[trees > 0] = 0


def mark_from_distance(prob_roof, max_dist):
    """
    Pixels that are too far from a 'positive' pixel are forced to
    be negatives.

    Parameters
    ----------
    :prob_roof: The probability that a pixel is rooftop
    :max_dist: The maximum distance a rooftop can be from the nearest
        shadow.

    """
    seeds = np.ones(prob_roof.shape, dtype=np.uint8)
    # NOTE: This is based on the max distance of a pixel from any pixel
    #      that has >90% score to be roof. It is not directly based on shadows.
    seeds[prob_roof >= 0.9] = 0
    distance = cv2.distanceTransform(seeds, cv2.cv.CV_DIST_L2, 3)
    out_of_range = distance >= max_dist
    prob_roof[out_of_range] = 0.0


def remove_large_and_small_buildings(buildings,
                                     prob_roof,
                                     min_area,
                                     max_area):
    """
    Removes contours from `building` whose area does not fall between a
    double threshold


    This keeps all contours where min_area <= countour area  <= max_area.
    Other contours are removed (set to 0).

    Parameters
    ----------
    :buildings: 1 if pixel is rooftop, else 0.
    :prob_roof: Between 0 and 1.
    :min_area: Lower thershold on contour area.
    :max_area: upper threshold on contour area.

    Returns
    -------
    A 2-tuple with the number that were too large, and the number
    that were too small.

    """
    contours, _ = cv2.findContours(buildings.astype(np.uint8),
                                   mode=cv2.RETR_EXTERNAL,
                                   method=cv2.CHAIN_APPROX_SIMPLE)

    large_contours = [c for c in contours if cv2.contourArea(c) > max_area]
    small_contours = [c for c in contours if cv2.contourArea(c) < min_area]

    if len(large_contours) > 0:
        cv2.drawContours(buildings, large_contours, -1, color=0, thickness=1)
        cv2.drawContours(prob_roof, large_contours, -1, color=0, thickness=1)
        cv2.drawContours(buildings, large_contours, -1, color=0, thickness=-1)
        cv2.drawContours(prob_roof, large_contours, -1, color=0, thickness=-1)

    if len(small_contours) > 0:
        cv2.drawContours(buildings, small_contours, -1, color=0, thickness=1)
        cv2.drawContours(prob_roof, small_contours, -1, color=0, thickness=1)
        cv2.drawContours(buildings, small_contours, -1, color=0, thickness=-1)
        cv2.drawContours(prob_roof, small_contours, -1, color=0, thickness=-1)

    info("Found %d %s", len(large_contours), "large contours...")
    info("    &  %d %s", len(small_contours), "small contours...")
    return len(large_contours), len(small_contours)


def remove_trees(image,
                 buildings,
                 prob_roof,
                 vegetation_threshold=0.2):
    """
    Remove 'buildings' that are too green to be buildings.

    This, in conjunction with the `mark_trees` function, is aimed at preventing
    trees from showing up as false positives.

    Paramters
    ---------
    :image:  Visible color RGB image
    :buildings:  Nonzero wherever we think a building is present
    :prob_roof:  Shades of gray indicating confidence

    Returns
    -------
    The number of buildings that were removed and marked as trees

    """

    mask = buildings.astype(np.uint8)
    contours, _ = cv2.findContours(mask,
                                   mode=cv2.RETR_EXTERNAL,
                                   method=cv2.CHAIN_APPROX_SIMPLE)

    trees = []
    for contour in contours:
        x0, y0, w, h = cv2.boundingRect(contour)
        x1, y1 = x0 + w, y0 + h
        mean = cv2.mean(image[y0:y1, x0:x1], mask=mask[y0:y1, x0:x1])

        if mean[1] > (1 + vegetation_threshold) * (mean[0] + mean[2]) / 2:
            trees.append(contour)

    # Black out the regions marked as green
    if len(trees):
        cv2.drawContours(buildings, trees, -1, color=0, thickness=1)
        cv2.drawContours(prob_roof, trees, -1, color=0, thickness=1)
        cv2.drawContours(buildings, trees, -1, color=0, thickness=-1)
        cv2.drawContours(prob_roof, trees, -1, color=0, thickness=-1)

    info("Found %d %s", len(trees), "probable trees ...")
    return len(trees)


```

```python
def grabcut(data,
            prob_roof,
            tmp1,
            tmp2,
            graphcut_iterations,
            use_median_blur=True):
    """
    Use grabcut to relax the probabilities determined from shadows so that
    the entire roof is pulled into the foreground while shadows remain in bg

    This routine converts the image to the LUV color space before running
    grabcut. It makes sure there are enough FGD and BGD constraints
    And it also records and logs the amount of time spend on gc

    """

    h, w = prob_roof.shape
    info("w,h= %d %d", w, h)

    if use_median_blur:
        data = cv2.medianBlur(data, 3)

    mask = cv2.GC_PR_BGD * np.ones((h, w), dtype=np.uint8)
    mask[prob_roof > 0.5] = cv2.GC_PR_FGD
    mask[prob_roof <= 0] = cv2.GC_BGD
    mask[prob_roof >= 1] = cv2.GC_FGD

    num_bgd = np.count_nonzero(prob_roof <= 0.5)
    num_fgd = np.count_nonzero(prob_roof > 0.5)

    info("BGD, FGD = %d %d", num_bgd, num_fgd)

    # Grabcut fails if there arent enough initial points
    if num_fgd <= 30 or num_bgd <= 30:
        info("Empty input, bailing out of grabcut")
        return

    luv = cv2.cvtColor(data, cv2.COLOR_RGB2LUV)
    luv -= np.min(luv)
    luv /= np.max(luv)
    data_8bit = np.clip(255 * luv, 0, 255).astype(np.ubyte)

    info("Calling cv2.grabCut with LUV image....")

    # The rect is ignored sinc we are using GC_INIT_WITH_MASK
    rect = (1, 1, w - 2, h - 2)
    cv2.grabCut(data_8bit,
                mask,
                rect,
                tmp1,
                tmp2,
                graphcut_iterations,
                mode=cv2.GC_INIT_WITH_MASK)
                
    if False:
       mask2 = np.array(mask)
       #mask2[mask==cv2.GC_BGD] = cv2.GC_PR_BGD
       mask2[mask==cv2.GC_FGD] = cv2.GC_PR_FGD
       cv2.grabCut( data_8bit,
                 mask2,
                 rect,
                 tmp1,
                 tmp2,
                 2,
                 mode=cv2.GC_INIT_WITH_MASK )
       mask2[(mask==cv2.GC_FGD)*(mask2==cv2.GC_PR_FGD)] = cv2.GC_FGD
       #mask2[(mask==cv2.GC_BGD)*(mask2==cv2.GC_PR_BGD)] = cv2.GC_BGD
       mask = mask2
       

    info("Writing grabcut labels back to the mask....")
    prob_roof[mask == cv2.GC_PR_BGD] = 0.25
    prob_roof[mask == cv2.GC_PR_FGD] = 0.75


def mark_non_shadow_contours(buildings,
                             prob_shadow,
                             light_vec,
                             max_overhang):
    """
    Mark pixels which should cast shadows but do not.

    """
    shadows = (prob_shadow == 0)
    shadow_structure = construct_structuring_element(2, light_vec)
    build_structure = construct_structuring_element(4, -1.0 * light_vec)

    suppose_shadow = scipy.ndimage.binary_dilation(
        buildings,
        structure=shadow_structure) & (
        ~buildings)
    non_shadow = suppose_shadow.astype(np.uint8) & (~shadows)
    non_shadow_build = scipy.ndimage.binary_dilation(
        non_shadow,
        structure=build_structure) & (
        ~non_shadow)

    prob_shadow[non_shadow_build > 0.5] = 0
    return non_shadow_build.sum()


def find_buildings_in_tile(settings,
                           image,
                           prob_roof=None,
                           loglevel=0,
                           recalculate=True):
    """
    Segment the buildings from a tile

    Parameters
    ----------
    :image:
            The visible color image (aerial photo), or a subimage
    :prob_roof:
            Mask that may have alreasy ben partially marked with BGD
            and FGD constraints

    :loglevel:
            The log level to use when initializing the logging system.
            The default value is 0.

    :recalculate=True:
            whether to calculate the shadow mask (`True`) or reuse an existin
            shadow mask (`False`).
            The default value is `True`.

    Returns
    -------
    :roof: ndarray.
        A raster array that is nonzero where a roof is visible
    :prob_roof:
        A raster array that is shades of gray indicating confidence that
        a pixel is roof

    """
    assert prob_roof is None or prob_roof.shape == image[:, :, 0].shape, \
        "Images must align pixel-to-pixel"

    logging.basicConfig(level=loglevel)

    shadow_correction_iterations = settings.get(
        'shadow_correction_iterations',
        2)
    light_vec = settings['light_vec']
    shadow_threshold = settings.get('shadow_threshold')
    ignore_thin_shadows = settings.get('ignore_thin_shadows', False)
    fgd_distance = settings['fgd_distance']
    max_overhang = settings.get('max_overhang', 7)
    tree_green_percent = settings.get('tree_green_percent', 25)
    graphcut_iterations = settings.get('graphcut_iterations', 10)
    min_building_area = settings.get('min_building_area', 600)
    max_building_area = settings.get('max_building_area', 512 * 512)

    # make sure grabcut settings are present
    tmp1 = settings['tmp1'] if 'tmp1' in settings else np.zeroes((1, 13 * 5))
    tmp2 = settings['tmp2'] if 'tmp2' in settings else np.zeroes((1, 13 * 5))


    R = image[:, :, 0]
    G = image[:, :, 1]
    B = image[:, :, 2]
    Yuv_Y = 0.2126 * R + 0.7152 * G + 0.0722 * B
    #Yuv_u = -0.09991*R - 0.33609*G + 0.436*B
    #Yuv_v = 0.615*R - 0.55861*G - 0.05639*B
    if image.max() == 0:
        return np.zeros(Yuv_Y.shape), np.zeros(Yuv_Y.shape)

    if recalculate:
        
        prob_roof = 0.5 * np.ones(Yuv_Y.shape)
        print "calculate shadows"
        mark_from_shadows(Yuv_Y, prob_roof, None,
                          light_vec,
                          shadow_threshold,
                          ignore_thin_shadows,
                          fgd_distance)
        mark_trees(image, prob_roof, tree_green_percent)

# TODO: We need to calculate this in meters!
#    info( "Eliminating pixels not near any shadow")
#    mark_from_distance( prob_roof,
#                        settings.get('max_building_diameter', 150))
    # Try grab-cut followed by shadow checks, for several
    # shadow_correction_iterations
    corrections = float("inf")
    buildings = (prob_roof > 0.5).astype(np.uint8)
    while shadow_correction_iterations > 0:
        info("Running grab-cut ... (This is the slow part) ....")
        grabcut(image, prob_roof, tmp1, tmp2, graphcut_iterations)
        buildings = (prob_roof > 0.5).astype(np.uint8)

        if shadow_correction_iterations > 1:
            corrections = mark_non_shadow_contours(prob_roof > 0.5,
                                                   prob_roof,
                                                   light_vec,
                                                   max_overhang)
            info("Identified {: 4} possible corrections.".format(corrections))
        shadow_correction_iterations -= 1


    remove_large_and_small_buildings(buildings,
                                     prob_roof,
                                     min_building_area,
                                     max_building_area)
    remove_trees(image, buildings, prob_roof)

    return buildings, prob_roof

```
